## A Simple Example of backtesting a strategy, on Monte Carle generated stock data ##


```python
# include imports from the directory above where systrade is
# could also install systrade in the pythonpath
import sys
sys.path.append('..')
# imports from systrade
from systrade.monte import generator
from systrade.monte import parameter as param
from systrade.monte import path
from systrade.monte import stats
from systrade.monte import stocks as stocksgen

from systrade.models import indicators
from systrade.models import signals
from systrade.models import strategies
from systrade.models import filters

from systrade.trading import brokers
from systrade.trading import accounts

from systrade.backtest import backtests

from systrade.plotting import plt_tseries

# imports from major libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Rather than pull financial data from real data, let's first generate a Monte Carlo set of prices.

We will do this for a set of two stocks - we will have to define their correlation, volatility, and the type of random motion that we expect them to follow. We will be using a simple geometric Brownian motion - with normally distributed log returns - we provide a normal number generator to our stock path generator for this reason.



```python
# make pathways for two stocks, we will call them benchmark and stock
stock_names = ['benchmark','stock']
# correlation of log-returns on two stocks
rho = np.ones((2,2))
rho[0,1] = 0.7
rho[1,0] = 0.7
# sigma, vol, of log stock returns
sig      = np.array([0.12,0.12])
# create generator that generates numbers on normal distribution
gen_norm = generator.Normal()
# create a stock pathway generator 
mystock_path_gen = stocksgen.MultiStockPath(gen_norm, rho, sig, stock_names)
```

```python
# define the first and last times we want to generate stock pathway over
t_start = pd.to_datetime('2019/07/15-09:30:00:000000', format='%Y/%m/%d-%H:%M:%S:%f')
t_end   = pd.to_datetime('2019/07/26-16:00:00:000000', format='%Y/%m/%d-%H:%M:%S:%f')

# interest rate in this period
interest_rate  = 0.024
# initial values of our two stocks (at t_start)
initial_values = [60,60]
# create a dataframe of the stocks over this time period
# we create values every 2 minutes
# we provide a specific random seed to the generator for reproducability
stocks_df = mystock_path_gen.get_path(initial_values,
                                      t_start,
                                      t_end,
                                      '2min',
                                      interest_rate,
                                      seed=0)
print(stocks_df.index.max())
```

Now we can start to build a trading strategy - to do this we begin by building some technical indicators - which we will in turn use to build trading signals.

One can easily build their own indicators - just have a class that inherits systrade.models.base.BaseIndicator, and provide a get_indicator method - that's all

```python
# create a systrade indicator that looks at moving average crossover
crossind = indicators.MACrossOver(int(0.2*6.5*60), int(0.6*6.5*60))
# create another systrade indicator that looks at moving average crossover
crossind2 = indicators.MACrossOver(int(1.0*6.5*60), int(2.0*6.5*60))
```

If we want to, we can look at what the indicator looks like on each ticker we have data for

```python
indi1_df = crossind.get_indicator(stocks_df)
indi2_df = crossind2.get_indicator(stocks_df)
```

Lets's now use our plotting tool to see how these indicators look over (market) time

```python

fig,axes = plt_tseries.subplots(stocks_df.index,1,1)

axes[0].plot_tseries(stocks_df.index,indi1_df['stock'],color='b',label='indicator 1: crossind')
axes[0].plot_tseries(stocks_df.index,indi2_df['stock'],color='m',label='indicator 2: crossind2')
axes[0].plot_horiz(0,stocks_df.index.min(),stocks_df.index.max(),color='k',LineStyle='--')

axes[0].set_ylabel('indicator value')
axes[0].set_xlabel('time (mins)')
axes[0].legend(frameon=False)

fig.autofmt_xdate()

plt.show()
```

Now, we'll want to create signals using our indicators.

To do so, we also need to create filter objects. These tell the signal how to take information from the indicator - which stocks to look at the indicator over, and which stocks to apply signals to on the basis of that indicator. 

That could be that we want to get the indicator over a single stock, and when that indicator does something specifc, say, cross zero, generate a signal to buy or sell that same stock. Alternatively, it may be that you want to get the indicator over a single stock, and on the basis of when that indicator does something, get a buy/sell signal on many other assets - e.g if the price of oil does something, we may want to buy/sell airline company stocks.

Here we will look at self-filtering - the signal will be to use indicator movements to get buy/sell signals on the same asset the indicator was generated on. Other options are available. 

```python
# create a systrade filter, that says that any indicator on 'stock', will give us info on whether to buy 'stock'
self_filter_stock = filters.TickerOneToAnotherFilter(['stock'],['stock'])

# create signals from our two indicators, that will send a buy signal if moving-average crossover goes through 0 upwards,
# and send a sell signal if the indicator goes through 0 downwards
MAsig = signals.ZeroCrossBuyUpSellDown(crossind,filter=self_filter_stock)
MAsig2 = signals.ZeroCrossBuyUpSellDown(crossind2,filter=self_filter_stock)
```

All signals provide a 'request_historical' method. This will generate a dictionary, with each key being a ticker, and the value for each key being a dataframe indexed by time of signals to buy/sell.

Again, users can create their own signal classes, inheriting systrade.models.base.BaseSignal, and providing a request_historical method - returning a dataframe indexed by time with +1 for a buy signal and -1 for a sell signal.

```python
sig_dict1 = MAsig.request_historical(stocks_df,signal_name='MA1')
sig_dict2 = MAsig2.request_historical(stocks_df,signal_name='MA2')
```

OK, let's look at what signals we got: +1 means a buy signal, -1 means a sell signal

```python
print("\n for first signal: \n:")
print(sig_dict1)
print("\n for second signal: \n:")
print(sig_dict2)
```

So for each signal, we get a dictionary, with the only key being 'stock' - as this was what we filtered our output (and indeed, input) on. And For stock, we get a dataframe of times with buy/sell (+/-1) signals

Lets plot out 'stock' along with the indicator, and the signals generated for the first case  

```python
#mydict={'bb':pd.DataFrame(data={'col':[1,2,3]})}
plt_tseries.plot_stock_indictor(stocks_df,'stock',indi1_df, signal_dict=sig_dict1)
```


Now that we've seen how to create indicators, signals, and seen what they look like, let's put that together into a trading strategy.

Trading strategies are formed of several signals. The strategy is given a dictionary of signals of creation - this (along with the software design) allows parameters of the strategy to be easily adjusted - as we'll see.

Strategies combine several signals, and apply any decisions based on their own parameters, to decide what orders to place. 

Let's make a SimpleStrategy, that simply adds up signals over 5 minute intervals - and buys/sell if overall signal is positive/negative. We will use our two signals above to create the strategy.

Again, users can create their own stratgies - just inheriting systrade.models.base.BaseSignal. Maybe you want to check if any of the orders already placed are now considered ricky. Maybe you want to also include methods for portfolio optimization and checking new signals don't interfere with the optimization too much - it's all possible.


```python
# create a strategy formed from these two signals, it will average their inputs to make a decision to buy/sell
mystrat = strategies.SimpleStrategy({'MA1':MAsig, 'MA2':MAsig2},['stock'],1.0)
```

We want to look at a range of parameters of the model, and asses how well the model performs for each case

We can do this by creating a backtest object, with a parameter grid to tell it which parameters we want to vary, and what values parameters should take - it will try all combinations of those parameters

Much like scikit-learn nested estimators (pipelines), parameters of the signals and indicators that make up a 
strategy can be accessed by using `__` between the nested objects.

A signal with name 'MA1' can have its own parameters adjusted via its name: 'MA1.param'.
The indictator of a signal named 'MA1' can have it's parameters adjusted using the name: 'MA1__indicator__param'


Backtesting also requires a broker. A broker object holds historical data, and will provide methods for retrieving parts of that data.

Further, an account factory is needed, this will generate an account with which to trade. We choose to create a BasicAccountFactory - which generates systrade.trading.accounts.BasicAccount objects.

The broker will provide different prices for buying and selling, have a delay between information requests and provided prices, just as for a real broker. These can be set on initialization of the broker.

Note that we could also initialize a broker using a dataframe of stock valuations that we obtained from real historical data from an API to a real world broker

```python

# create grid of parameters to be adjusted: a dict with each parameter name having a list of values
params = {'MA1__indicator__period1':[70,80],
          'MA1__indicator__period2':[200,300],
          'MA2__indicator__period1':[100,200],
          'MA2__indicator__period2':[300,400]
         }

# params = {'MA1__indicator__period1':[70,70],
#           'MA1__indicator__period2':[200,200],
#           'MA2__indicator__period1':[100,100],
#           'MA2__indicator__period2':[300,300]
#          }




# create a broker object holding the historical data - the broker provides methods for getting market prices.
mybroker  = brokers.PaperBroker(stocks_df)

# create an account factory - that will build accounts on whcih to trade - use a basic account
account_factory = accounts.BasicAccountFactory()

# create the ParameterScanBackTest, using our broker, strategy and parameter grid
backtester = backtests.ParameterScanBackTest(mybroker,account_factory,mystrat,params)
```

Now we have created our backtest object, we need to run it and see how well the strategy performs.

Determining how well a strategy performs is a statistical problem. We won't get into too much detail here, but note that the p-value of a strategy over the test is in indication of the statistical likelihood of that result being observed by chance - and we'd like it to be low. When multiple tests are run - as here - the p-value is no longer appropriate. One must consider that when testing many strategies we increase our chances of doing well just by chance, and so one can consider the family-wise error rate. There are different ways to calculate adjusted p-values, and two are available in the backtester. We set the significance here to 0.35, although in real-trading situations one would want a much lower fwer significance, around 0.05.

```python
# now run the backtest, using the benchmark stock as the benchmark, and setting the family-wise error rate to a high 0.35
adj_p = backtester.run_bootstrap_all(stocks_df['benchmark'],fwer_alpha=0.35)
```

```python
# we can get all the results of the test in a dataframe format
# optionally included in the dataframe a dictionary of the parameters used in that strategy
print(backtester.get_results_df(include_params=True))
```

```python
# we can see which of the strategies that were tried in the parameter scan backtest passed the FWER test
good_strategy_list = backtester.get_successful_strategies()
print(good_strategy_list)
```

```python
# we can also access the test data of those tests that passed (i.e for which we could reject the Null hypothesis)
good_strategy_data_list = backtester.get_successful_strategies_and_data()
# this list contains a tuple for each succesful strategy: (the strategy itself, the data of the test)
good_strat0      = good_strategy_data_list[0][0]
good_strat0_data = good_strategy_data_list[0][1]

print(good_strat0.get_params())
print(good_strat0_data.adjusted_p_value)
print(good_strat0_data.null_rejected)
print("\n\n vars of the test are \n\n:")
print(vars(good_strat0_data))

```
